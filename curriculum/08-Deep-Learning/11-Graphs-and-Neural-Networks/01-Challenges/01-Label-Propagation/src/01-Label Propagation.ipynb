Here's a reminder of the Label propagation algorithm:

At step 0, define:

$ F(1) = \alpha S Y + (1-\alpha) Y $

Where:
- $S$ is an adjacency matrix, and shows the relations in the graph:

$$
S = \begin{pmatrix}
0 & 1 & 1\\
1 & 0 & 0\\
1 & 0 & 0
\end{pmatrix}
$$

- $Y$ is the initial label (one-hot encoded):

$$
Y = \begin{pmatrix}
0.5 & 0.5\\
0. & 1\\
1 & 0
\end{pmatrix}
$$

- $ F(t) $ on its side denotes the probability that each node belongs to each label, determined iteratively, for example:
    
$$
F(t) = \begin{pmatrix}
0.5 & 0.5\\
0.2 & 0.8\\
0.6 & 0.4
\end{pmatrix}
$$

Then, iterate for a fixed given number of times: $ F(t+1) = \alpha S F(t) + (1-\alpha) Y $
